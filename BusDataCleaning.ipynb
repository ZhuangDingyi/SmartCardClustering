```python
html=urllib.request.urlopen("https://www.transitlink.com.sg/eservice/eguide/bscode_idx.php?bs_code=59619")
html_doc=html.read()

soup = BeautifulSoup(html_doc)
bus_codes_list=soup.find_all('option')
bus_codes=[]
#Fetch all the bus codes
#a=soup.find_all(class_='data')
#soup.select(td[class_="data"])
#soup("td",class_="data")
for i in bus_codes_list:
    bus_codes.append(i.get('value'))
```

```python
def getBusStopName(BusCode):
    html_='https://www.transitlink.com.sg/eservice/eguide/bscode_idx.php?bs_code='
    html_url=html_+BusCode
    html=urllib.request.urlopen(html_url)
    html_doc=html.read()
    soup = BeautifulSoup(html_doc)
    data=soup.find_all(class_="data")
    busStopName=data[0].get_text()
    return busStopName
```

```python
busStopNames=[]
```

```python
for i in range(len(bus_codes)):
    print(i/len(bus_codes))
    if len(bus_codes[i])>3:
        busStopNames.append(getBusStopName(bus_codes[i]))
    else:
        del bus_codes[i]
```
